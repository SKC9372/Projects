```python
# Importing libraries for creating model
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score

import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Splitting data into train and test data 
df_train , df_test = train_test_split(boom , train_size = 0.7 , random_state = 100)
```

```python
df_train.shape
```

```python
# Normalizing the train data for better convergence
numeric = ['temp','atemp' , 'hum' , 'windspeed' ,  'cnt']
scaler = MinMaxScaler()
df_train[numeric] = scaler.fit_transform(df_train[numeric])
```

```python
df_train.describe()
```

```python
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
X_train.shape
```

```python
def vif(X):
    vif = pd.DataFrame()
    vif['feature'] = X.columns
    vif['VIF']  = [variance_inflation_factor(X.values , i) for i in range(X.shape[1])]
    vif['VIF']  = round(vif['VIF'] , 2)
    vif = vif.sort_values(by = 'VIF' , ascending = False)
    return vif
```

Creating Models 

```python
X_train_sm = sm.add_constant(X_train)
lm_1 = sm.OLS(y_train , X_train_sm).fit()
lm_1.summary()
```

```python
vif(X_train_sm.drop('const' , axis = 1))
```

```python
X_train_sm = sm.add_constant(X_train.drop('mnth', axis = 1))
lm_2 = sm.OLS(y_train , X_train_sm).fit()
lm_2.summary()
```

```python
vif(X_train_sm.drop('const' , axis = 1))
```

```python
X_train_sm = sm.add_constant(X_train.drop(['mnth' ,'atemp' ] , axis = 1))
lm_3 = sm.OLS(y_train , X_train_sm).fit()
lm_3.summary()
```

```python
vif(X_train_sm.drop('const' , axis = 1))
```

```python
X_train_sm = sm.add_constant(X_train.drop(['mnth' ,'atemp', 'weekday'] , axis = 1))
lm_4 = sm.OLS(y_train , X_train_sm).fit()
lm_4.summary()
```

```python
vif(X_train_sm.drop('const' , axis = 1))
```

```python
X_train_sm = sm.add_constant(X_train.drop(['mnth' ,'atemp', 'weekday','hum'] , axis = 1))
lm_5 = sm.OLS(y_train , X_train_sm).fit()
lm_5.summary()
```

```python
vif(X_train_sm.drop('const' , axis = 1))
```

```python
y_train_pred = lm_5.predict(X_train_sm)
```

```python
res = y_train - y_train_pred
```

```python
#Evaluating error terms 
sns.distplot(res)
plt.title('Residual Distribution Analysis' , color = 'red')
plt.show()
```

Error terms are normally distributed showing there mean is 0 and follow homoscedasticity

Linear Regression using RFE

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train , y_train)
rfe = RFE(estimator = lr , n_features_to_select = 10)
rfe = rfe.fit(X_train , y_train)
```

```python
list(zip(X_train.columns , rfe.support_ , rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = sm.add_constant(X_train[col])
lr_1 = sm.OLS(y_train , X_train_rfe).fit()
lr_1.summary()
```

```python
vif(X_train_rfe.drop('const' , axis = 1))
```

```python
# Creating the final model
X_train_new = X_train_rfe.drop('hum' , axis = 1)
X_train_new = sm.add_constant(X_train_new)
lr_2 = sm.OLS(y_train , X_train_new).fit()
lr_2.summary()
```

```python
vif(X_train_new.drop('const' , axis = 1))
```

```python
y_predict = lr_2.predict(X_train_new)
```

```python
r2_score(y_true = y_train , y_pred = y_predict)
```
